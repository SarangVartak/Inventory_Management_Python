```python
import json


def showinventory():
  a = open("inventory.json", "r")
  dat = json.load(a)
  a.close()
  print ("Product Id","\t", "Name", "\t\t", "Price", "\t", "Quantity", "\t", "Discount%", "\t" ,"ExpiryDate")
  print("*******************************************************************************************************************************")
  for i in dat:
    print( i,"\t", dat[i]['Name'], "\t", dat[i]['Rate'], "\t", dat[i]['Quantity'], "\t\t", dat[i]['Discount'], "\t\t", dat[i]['Expiry'])
  print("*******************************************************************************************************************************")



def showbill(abill) :
  import time
  currtime =  time.ctime()
  print("Transaction Successful")
  print(currtime)
  print("Product Id", "\t", "Name", "\t\t", "Rate", "\t", "Quantity", "\t", "Amount(Discounted)")
  print("*******************************************************************************************************************************")
  total = 0 ;
  for eac in abill:
    print( eac['pid'], "\t", eac['name'], "\t", eac['rate'], "\t", eac['quantity'], "\t\t", eac['amount'] )
    total += eac['amount']
  print ("Total  : ", total)
  print("*******************************************************************************************************************************")
  print("Thanks for shopping with us!! ")
  import json
  tr = open("transactions.json", "w")
  trans = {"Date" : currtime, "items" : abill, "total" : total}
  tr.write(json.dumps(trans, indent = 4))
  tr.close()


def placeorder():
  a = open("inventory.json", "r")
  dat = json.load(a)
  a.close()
  nitems = int(input("Enter the Number of Products you want : "))
  itemlis = []
  for i in range(nitems) :
    print("Product Number ", i + 1)
    pid = int(input("Enter Product Id : "))
    quan = int(input("Enter the Quantity : "))
    itemlis.append({"pid" : pid, "quan" : quan})
  bill = []
  for it in itemlis:
    a = dat[str(it['pid'])]
    if a['Quantity'] >= it['quan'] :
      dat[str(it['pid'])]['Quantity'] = a['Quantity'] - it['quan']
      dis = a['Discount']
      dis = 100 - dis
      dis = dis / 100.0
      bill.append({"pid" : it['pid'], "name" : dat[str(it['pid'])]['Name'], "rate" : dat[str(it['pid'])]['Rate'], "quantity" : it['quan'] ,"amount" : (dat[str(it['pid'])]['Rate'] * it['quan'] * dis) })
  fd = open("inventory.json", "w")
  fd.write(json.dumps(dat, indent = 4))
  fd.close()
  showbill(bill)
  pass


ch = 1 ;
while(ch) :
  print("Please choose any one of the following")
  print("1. Inventory")
  print("2. Place Order")
  print("0. Exit")
  ch = int(input("Provide input here : "))
  if ch == 1 :
    showinventory()
  elif ch == 2 :
    placeorder()
    ch = 0
  elif ch == 0 :
    print("Exiting, Have a Nice Day")
  else :
    print("Wrong choice, Exiting!!")
    ch = 0
```
